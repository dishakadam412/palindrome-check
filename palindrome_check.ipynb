```python
# palindrome-check

s = input("Enter word : ") ## store user input in variable

## print(s)

reverse = s[::-1] ## take user input, reverse it, store it in a new variable

## print(reverse)

if(s == reverse):
    print("This word is a palindrome!") ## If the original input and reverse input match, then the word is a palindrome
else:
    print("This word is not a palindrome") ## If the original input and reverse input dont match, then the word is not a palindrome
```
